```python
#HALNDLING MISSING VALUES
```

```python
airbnb.isnull().sum()
```

```python
airbnb["host_name"] = airbnb["host_name"].fillna("Unknown")
```

```python
airbnb["reviews_per_month"] = airbnb["reviews_per_month"].fillna(
    airbnb["reviews_per_month"].median()
)
```

```python
airbnb["last_review"] = airbnb["last_review"].fillna("No Review")
```

```python
airbnb.isnull().sum()
```

```python
airbnb.duplicated().sum()
```

```python
airbnb = airbnb.drop_duplicates()
```

```python
airbnb.columns = airbnb.columns.str.lower()

airbnb.columns = airbnb.columns.str.replace(" ", "_")
```

```python
airbnb.columns
```

```python
#CHECKING OUTLIERS
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))

sns.boxplot(x=airbnb["price"])

plt.title("Price Outlier Detection")

plt.show()
```

```python
#REMOVING OUTLIERS
```

```python
q1 = airbnb["price"].quantile(0.25)

q3 = airbnb["price"].quantile(0.75)

iqr = q3 - q1

lower = q1 - 1.5*iqr

upper = q3 + 1.5*iqr

airbnb = airbnb[
    (airbnb["price"] >= lower)
    &
    (airbnb["price"] <= upper)
]
```
